# Creating instances of the Moving Firefighter Problem

```python
import networkx as nx
import numpy as np

import movingfp.gen as mfp

np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})
```

## 2D Connected Erdos instance

```python
n = 8
p = 0.5
dim =  2
fighter_pos = None
num_fires =  2

seed = None
generator = np.random.default_rng(seed)

x = mfp.erdos_connected(n, p, dim, fighter_pos, num_fires, generator)
mfp.plot2d(x)
```

```python
# Adjacency matrix (Fire graph)
x.A
```

```python
# Distance matrix (Firefighter graph)
# Last row and column are firefighter distances to every node
x.D
```

## 3D Connected Erdos Instance

```python
n = 15
p = 0.25
dim =  3
fighter_pos = None
num_fires =  2
seed = None
generator = np.random.default_rng(seed)

x = mfp.erdos_connected(n, p, dim, fighter_pos, num_fires, generator)
mfp.plot3d(x, node_size=7, plot_grid=True, plot_labels=True)
```
